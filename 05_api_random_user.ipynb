```python
from randomuser import RandomUser
import pandas as pd
```

```python
r = RandomUser()
```

```python
some_users = r.generate_users(10)
some_users
```

```python
some_users[1].get_full_name()
```

```python
for user_ in some_users:
    # print(user_.get_full_name())
    print(f"user is {user_.get_full_name()} and email is {user_.get_email()}")
```

```python
for user_ in some_users:
    print(f"photo is {user_.get_picture()}")
```

```python
def get_users(num_users):
    # users_list will hold a list of dictionary with users details
    users_list = [] 

    for users_ in RandomUser().generate_users(num_users):
        users_list.append({ "id": users_.get_id(), \
                            "city": users_.get_city(), \
                            "name": users_.get_full_name() , \
                            "age": users_.get_age() })

    return pd.DataFrame(users_list)

print(get_users(20))

```
